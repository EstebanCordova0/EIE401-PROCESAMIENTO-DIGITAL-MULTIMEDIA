### 3.2 Genere una señal Senoidal y realice la operación de sampleo
Puedes utilizar el método "stem" de matplotlib para plotear el resultado.

```python
# Señal creada
amplitud = 4
frecuencia = 10
f_muestreo = [2,20,21,60,200]
tiempo = 1

# Tiempo señal continua y muestreada
tiempo_continuo=np.linspace(0, 1/frecuencia, 200)
tiempo_sampleo = np.arange(0, 1 , 1/f_muestreo[2])

# Señal continua
continua = amplitud * np.sin(2 * np.pi * frecuencia * tiempo_continuo)


# Grafica señal continua
plt.figure(figsize=(10, 5))
plt.subplot(2,1,1)
plt.plot(tiempo_continuo, continua, label="Señal continua")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Señal continua")
plt.legend()
plt.grid()

# Grafica señal muestreada
for i in range(len(f_muestreo)):
    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(0, tiempo, 1/f_muestreo[i]), amplitud * np.sin(2 * np.pi * frecuencia * np.arange(0, tiempo, 1/f_muestreo[i])))
    plt.stem(np.arange(0, tiempo, 1/f_muestreo[i]), amplitud * np.sin(2 * np.pi * frecuencia * np.arange(0, tiempo, 1/f_muestreo[i])))
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.title(f"Sampleo de una señal sinusoidal con frecuencia {f_muestreo[i]} de muestreo")
    plt.grid()
    plt.show()
```

### 3.3 Genere una señal Triangular y realice la operación de sampleo

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
amplitud = 4
frecuencia = 10  # Hz
tiempo = 1       # segundos
f_muestreo = [2,20,21,60,200]
tiempo_continuo=np.linspace(0, 1/frecuencia, 200)

# Generar señal triangular

triangular = (2 * amplitud / np.pi) * np.arcsin(np.sin(2 * np.pi * frecuencia * tiempo_continuo))


# Grafica señal triangular
plt.plot(tiempo_continuo, triangular, label="Triangular", color='orange')
plt.title("Señal Triangular")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid()
plt.legend()

# Grafica señal muestreada
for i in range(len(f_muestreo)):
    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(0, tiempo, 1/f_muestreo[i]), (2 * amplitud / np.pi) * np.arcsin(np.sin(2 * np.pi * frecuencia * np.arange(0, tiempo, 1/f_muestreo[i]))))
    plt.stem(np.arange(0, tiempo, 1/f_muestreo[i]), (2 * amplitud / np.pi) * np.arcsin(np.sin(2 * np.pi * frecuencia * np.arange(0, tiempo, 1/f_muestreo[i]))))

    plt.xlabel("Tiempo (s)")
    plt.ylabel("Amplitud")
    plt.title(f"Sampleo de una señal sinusoidal con frecuencia {f_muestreo[i]} de muestreo")
    plt.grid()
    plt.show()
```

### 3.5 Compruebe el teorema de Nyquist
Para comprobarlo, debes evaluar algunos casos:
1. la frecuencia de sampleo _f_ es mayor que la frecuencia máxima de tu señal, Fmax.
2. la frecuencia de sampleo _f_ es mayor que 2*Fmax.
3. la frecuencia de sampleo _f_ es mayor que 5*Fmax.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
f = 10  # Hz
duracion = 1  # segundo
fs_continua = 1000  # Alta frecuencia para simular señal continua

# Tiempo y señal original
t_continua = np.linspace(0, duracion, fs_continua)
señal_original = np.cos(2*np.pi*f*t_continua)+np.sin(np.pi*f*t_continua+3)

# Grafica señal original
plt.figure(figsize=(10, 4))
plt.plot(t_continua, señal_original, label='Señal original', color='blue')
plt.title('Señal Original (Continuo)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import resample

# Parámetros base
f = 10  # Hz
fmax = 2 * f  # 20 Hz
duracion = 1  # segundo
fs_continua = 1000

# frecuencias de muestreo
f_muestreo = [25, 45, 120]

for i, fs in enumerate(f_muestreo):
    # Señal original continua
    t_continua = np.linspace(0, duracion, fs_continua)
    señal_original = np.cos(2*np.pi*f*t_continua)+np.sin(np.pi*f*t_continua+3)

    # Señal sampleada
    t_sampleado = np.arange(0, duracion, 1/fs)
    señal_sampleada = np.cos(2*np.pi*f*t_sampleado)+np.sin(np.pi*f*t_sampleado+3)

    # Reconstrucción
    señal_reconstruida = resample(señal_sampleada, len(t_continua))

    # Grafica Señales
    plt.figure(figsize=(10, 4))
    plt.plot(t_continua, señal_original, label='Original', color='gray')
    plt.stem(t_sampleado, señal_sampleada, basefmt=' ', linefmt='r-', markerfmt='ro', label='Sampleada')
    plt.plot(t_continua, señal_reconstruida, 'g--', label='Reconstruida')
    plt.title(f'Caso {i+1}: f_s = {fs} Hz')
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.grid()
    plt.legend()

    plt.show()

```

### 4.3 Obtenga la transformada de Fourier para las dos señales que generaste previamente


```python

import numpy as np
import matplotlib.pyplot as plt

# Señal 1
x = np.linspace(0, 10, 100)
señal_1 = np.sin(x) + np.cos(2 * x)
N = len(señal_1)

#señal 2
f = 10
steps = 0.001
npTime = np.arange(0, 0.5, step=steps)
señal = np.sin(2 * np.pi * f * npTime) * np.cos(2 * np.pi * f * npTime + 4)
ruido_gaussiano = np.random.normal(0, 0.2, npTime.shape)
señal_2 = señal + ruido_gaussiano


# Transformada de fourier
def Transformada_fourier(señal1):
    N = len(señal1)
    X = []
    for k in range(N):
        sumatoria = 0
        for n in range(N):
            sumatoria += señal1[n] * np.exp(-2j * np.pi * k * n / N)
        X.append(sumatoria)
    return np.array(X)

# Aplicar transformada de fourier
espectro_transformada = Transformada_fourier(señal_1)
magnitude_transformada = np.abs(espectro_transformada) / N
frecuencias = np.fft.fftfreq(N, d=(x[1] - x[0]))  # solo para generar el eje x

# Aplicar transformada de fourier 2
N = len(señal_2)
espectro_transformada_2 = Transformada_fourier(señal_2)
magnitude_transformada_2 = np.abs(espectro_transformada_2) / N
frecuencias_2 = np.fft.fftfreq(N, d=(x[1] - x[0]))  # solo para generar el eje x

# Graficas señales originales y señales recuperadas
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x, señal_1, label='Señal Original señal 1')
plt.title('Señal Original señal 1')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(2, 2, 2)
plt.stem(frecuencias, magnitude_transformada)
plt.title("Transformada de Fourier (DFT manual) - Señal 1")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud (Normalizada)")
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(npTime, señal_2, label='Señal Original señal 2')
plt.title('Señal Original señal 2')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(2, 2, 4)
plt.stem(frecuencias_2, magnitude_transformada_2)
plt.title("Transformada de Fourier (DFT manual) - Señal 1")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud (Normalizada)")
plt.grid()
plt.tight_layout()
plt.show()
```

### 4.5 Encuentra la transformada inversa de Fourier
_Solo para una señal de las dos que generaste inicialmente_.

En este caso puede usar una librería.

Explica en detalle que hace la librería y el resultado que arroja.

```python
import numpy as np
import matplotlib.pyplot as plt

# Señal 1
x = np.linspace(0, 10, 100)
señal = np.sin(x) + np.cos(2 * x)
N = len(señal)

# Transformada de fourier
def Transformada_fourier(señal1):
    N = len(señal1)
    X = []
    for k in range(N):
        sumatoria = 0
        for n in range(N):
            sumatoria += señal1[n] * np.exp(-2j * np.pi * k * n / N)
        X.append(sumatoria)
    return np.array(X)

# Aplicar transforma a la señal
espectro_transformada = Transformada_fourier(señal)
frecuencias = np.fft.fftfreq(N, d=(x[1] - x[0]))  # solo para generar el eje x

# Usar la función de numpy para la inversa
señal_recuperada_numpy = np.fft.ifft(espectro_transformada)

# Graficar señal recuperada
plt.figure(figsize=(12, 6))
plt.plot(x, señal_recuperada_numpy, label='Señal Recuperada', color='r', linestyle='--')
plt.title('Señal Recuperada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid()
plt.show()
```

La funcion de la libreria numpy toma una señal en el dominio de la freceuncia y devuelve su representacion en el dominio de tiempo, en este caso uso la transforamada de fourier para reconstruirla. Finalmente el resultado es una señal muy parecida a la original

Hallazgos:


*   Si la frecuencia de muestro es menor al doble de la frecuencia maxima esta no se puede reconstruir correctamente , esto se puede comprobar por el teorema de nyquist, ademas que una alta frecuencia de muestreo logra una reconstruccion mas precisa.

*   El ruido gaussiano es de tipo aleatorio que sigue una distribucion normal, este se llama ruido blanco porque contiene todas las frecuencias con igual potencia, el efecto en la onda original es producir pertubarciones suaves y continuas ademas afecta por igual a todas las frecuencias de la señal.El ruido uniforme tiene una distribucion uniforme continua esto quiere decir cada valor dentro de un intervalo tiene la misma probabilidad de ocurrir, el efecto en la onda original es producir perturbaciones contantes dentro de un rango fijo.

*   La transformada de fourier descompone la señal en su frecuencia, esto se refiere a que frecuencias estan en la señal y con que intensidad, en el primer caso de la señal 1 se puede observar que esta va de una frecuencia de -5 Hz a 5 Hz con una amplitud maxima entre las frecuencias centrales esto esta correcto ya que en la señal 1 original se puede ver que tiene una amplitud mayor en la zona central.En la segunda señal o señal 2 se puede observar que la intensidad de la señal tiene cambios muy suaves en toda la frecuencia, pero la frecuencia de esta señal abarca desde los 0 Hz hasta los 500 Hz, esto esta correcto ya que la señal es de forma gaussiana lo que quiere decir que abarca muchas frecuencias y ademas la amplitud es muy poco cambiante como se puede observar en el ejemplo.
