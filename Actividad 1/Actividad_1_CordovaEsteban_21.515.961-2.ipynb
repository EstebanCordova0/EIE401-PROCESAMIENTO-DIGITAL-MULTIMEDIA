### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
import numpy as np
import matplotlib.pyplot as plt

# Señal sinusoidal
f = 10  
steps = 0.001  
npTime = np.arange(start=0,stop=0.5,step=steps)
señal = np.sin(2 * np.pi * f * npTime)

# Señal sinusoidal con ruido gaussiano
ruido_gaussiano = np.random.normal(0, 0.2, npTime.shape)
señal_con_ruido = señal + ruido_gaussiano

# Grafico de las señales
plt.figure(figsize=(10, 6))
plt.plot(npTime, señal)
plt.plot(npTime, señal_con_ruido)
plt.title("Señal con ruido gaussiano")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido rosa es un tipo de ruido cuyo espectro de potencia es inversamente proporcional a la frecuencia. A diferencia del ruido gaussiano el cual muestra fluctuaciones aleatorias que se distribuyen normalmente alrededor de la señal original, la densidad espectral de potencia del ruido rosa disminuye a medida que aumenta la frecuencia.
    Esto se puede observar en el grafico siguiente.

```python
import numpy as np
import matplotlib.pyplot as plt

# Señal sinusoidal
f = 10  
steps = 0.001  
npTime = np.arange(start=0,stop=0.5,step=steps)
señal = np.sin(2 * np.pi * f * npTime)


# Funcion ruido
def noise_psd(N, psd = lambda f: 1):
        X_white = np.fft.rfft(np.random.randn(N));
        S = psd(np.fft.rfftfreq(N))
        # Normalize S
        S = S / np.sqrt(np.mean(S**2))
        X_shaped = X_white * S;
        return np.fft.irfft(X_shaped);

# Funcion para generar ruido aleatorio
def PSDGenerator(f):
    return lambda N: noise_psd(N, f)

# Funcion ruido rosa
@PSDGenerator
def pink_noise(f):
    return 1/np.where(f == 0, float('inf'), np.sqrt(f))


# Señal ruido rosa 
ruido_pink = pink_noise(len(npTime))
señal_con_ruido_rosa = señal + ruido_pink

# Grafica de las señales
plt.figure(figsize=(10, 6))
plt.plot(npTime, señal)
plt.plot(npTime, señal_con_ruido_rosa)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.title("Señal con ruido rosa")
```
